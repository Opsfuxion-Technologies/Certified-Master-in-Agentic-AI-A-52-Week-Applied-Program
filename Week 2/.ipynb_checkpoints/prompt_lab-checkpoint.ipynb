### Base GPT call

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
 
load_dotenv()
client = OpenAI()
 
def ask_gpt(prompt):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role":"user","content":prompt}]
    )
    return response.choices[0].message.content
 
print(ask_gpt("Hello, who are you?"))
```

### 2. Zero-Shot Prompting

```python
print(ask_gpt("Summarize the plot of the movie, Inception in 3 sentences."))
## Observe clarity but possible vagueness.
```

### 3. Few-Shot Prompting

```python
# Give examples in the prompt:

prompt = """
Translate English to Pirate Speak:
English: Hello friend
Pirate: Ahoy matey
 
English: Where is the treasure?
Pirate:"""
print(ask_gpt(prompt))

prompt1 = """
Translate English to Arabic:
English: Hello friend, My name is Amara
I am from Nigeria and would love to travel the world,
meet new people.. learn new cultures,
have fun and be educated.""" 

print(ask_gpt(prompt1))

prompt2 = """
Translate English to Russian:
English: Hello friend, My name is Amara
I am from Nigeria and would love to travel the world,
meet new people.. learn new cultures,
have fun and be educated.""" 

print(ask_gpt(prompt2))

# Add at least 2 examples and test different completions.
```

### 5. Role Prompting (10 min)

```python
prompt = """
You are a helpful math tutor. 
Explain how to solve 16 √∑ 4 to a 10-year-old.
"""
print(ask_gpt(prompt))

prompt01 = """
You are a military drill sergeant. 
Explain how to solve 16 √∑ 4 to a new recruit.
"""
print(ask_gpt(prompt01))

prompt02 = """
You are a helpful comedian. 
Explain how to solve 16 √∑ 4 to a new recruit.
"""
print(ask_gpt(prompt02))

# Change the role to ‚Äúprofessor,‚Äù ‚Äúcomedian,‚Äù ‚Äúmilitary drill sergeant‚Äù and compare tone.
```

### 6. Instruction vs Open-Ended (10 min)

```python
print(ask_gpt("List 5 cities in Europe."))
print(ask_gpt("Tell me about Europe."))
# Compare structured vs vague instructions.
```

### 7. Chain-of-Thought (COT) 
#### Prompting Encourage reasoning:

```python
prompt = """
Solve: A train leaves at 3pm traveling 60 mph. Another leaves at 4pm at 90 mph. 
When will the faster train catch up? 
Think step by step.
"""
print(ask_gpt(prompt))
# Note intermediate reasoning.
```

### 8. Adversarial / Ambiguous Prompts 

```python
print(ask_gpt("Write a recipe for success."))
print(ask_gpt("Write a recipe for baking a cake in JSON format."))
# Observe how format constraints change output.
```
