```python
# JSON & API Basics (20 min)
import json
 
# JSON-like structure
agent_data = {
    "agent": "Alpha",
    "role": "assistant",
    "goals": ["plan", "analyze", "act"]
}
 
# Convert to JSON string
json_str = json.dumps(agent_data)
print("JSON string:", json_str)
 
# Back to Python dict
parsed = json.loads(json_str)
print("Agent goals:", parsed["goals"])

# If you want to stretch:
```
